### Big O Notation

Big O notation is a way to describe the performance of an algorithm, and is a way to compare an algorithms efficiency to other algorithms.

For example, suppose we have an algorithm that takes in a list of integers, and we want to know how the runtime of the algorithm scales as the size of the input list grows. We use Big O notation to describe this scaling;




Some common time complexities in Big O notation are:
- O(1) : Constant complexity. Number of steps is independent of the size of the input.
- O(n) : Linear complexity. Number of steps is directly proportional to the size of the input.
- O(n^2) : Quadratic complexity. Number of steps is proportional to the square of the size of the input.
- O(n!) : Factorial complexity. Number of steps is proportional to the factorial of the size of the input.

### Examples:

Lets look at a few examples to see how Big O notation works in practice.
Below we have a function that returns the first element in a list:

```python
def first_element(lst):
    return lst[0]
```

Since the function returns the first element of the list, the number of steps is independent of the size of the input. Meaning that the complexity of this function is O(1).

Now we have a function that sums up all the elements in a list:

```python
def sum_elements(lst):
    total = 0
    for x in lst:
        total += x
    return total
```

The number of steps is directly proportional to the size of the input list, because the loop iterates over all the elements in the list. Therefore, the complexity of this function is O(n).

Already we are able to observe the vast differences than can occur with these Time complexities. When working with bigger and bigger amounts of data these become increasingly relevant.

### Growth rates

Computational growth rates: the running time of an algorithm grows as the input size increases. These rates can be used to compare the efficiency of different algorithms, and are necessary when working with large amounts of data.

To explain growth rates we must use our knowledge of complexity:

```python
# Numbers.
import numpy as np

# Plotting.
import matplotlib.pyplot as plt

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# x values.
x = np.linspace(0.0, 10, 1000)

# Plot polynomial.
ax.plot(x, x**2, label="$x^2$")

# Plot exponential.
ax.plot(x, 2**x, label="$2^x$")

# Legend.
ax.legend()

# Show.
plt.show();
```

It's important to note the input size and the required accuracy of the result: For small inputs, the difference in running time between two algorithms with different time complexities may not be as noticeable. However, for large inputs the difference can be more significant.

### Turing machine


The Turing machine (created by Alan Turing, The grandfather of computers) is a machine that was designed to solving compuational problems in computer science and testing the limits of computation. Below is a demonstration of how it works mathematically using python:

```python
# State table.
table = {
    ('X', '_'): ['_', 'R', 'T'],
    ('X', '0'): ['0', 'R', 'X'],
    ('X', '1'): ['1', 'R', 'Y'],
    ('Y', '_'): ['_', 'R', 'F'],
    ('Y', '0'): ['0', 'R', 'Y'],
    ('Y', '1'): ['1', 'R', 'X'],
}

# Tape input.
tape = list('0110011')
# Position on tape.
pos = 0
# Initial state is first in table.
state = 'X'

# Keep going while we are not in a halting state.
while state not in ['T', 'F']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.
    row = table[(state, tape[pos])]
    # Overwrite the symbol.
    tape[pos] = row[0]
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        # Decrease position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```
