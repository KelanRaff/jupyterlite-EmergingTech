# Quantum Computing

**What is Quantum Computing?**

Quantum computing is a theory of computing that uses quantum-mechanics, such as **superposition and entanglement**, to perform computational operations. It has the potential to solve certain problems much faster than classical computers, due to the principles of quantum mechanics allowing quantum computers to execute certain calculations simultaneously.

In a computer, data is organised into bits, which is represented to us as binary.

In a quantum computer, data is organised into quantum bits, or qubits. Qubits can represent a 0, a 1, or both values simultaneously (**quantum super-positioning**)

### Flipping a Coin


The probability of flipping a coin and getting a specific result, heads or tails, is always 1 out of 2, or 1/2, or 50%. This is because the coin has two sides, and there are only possibilities when flipping it: heads or tails. The probability of any given outcome is the number of ways that event can occur divided by the total number of possible outcomes.

In the case of tossing a coin, there is only 1 way to flip heads and 1 way to flip tails, and there are 2 total possible outcomes, so the probability of flipping heads or tails is always 50%.

```python
# Generating random numbers.
import numpy as np

# Plotting.
import seaborn as sns

# Random numbers and choices.
import random

# Combinations and permutation.
import itertools
```

```python
# Flip one coin.
np.random.binomial(1, 0.5)
```

```python
# Flip twenty coins.
np.random.binomial(1, 0.5, 20)
```

```python
# Flip a coin one hundred times, counting total heads.
np.random.binomial(100, 0.5)
```

```python
# Should be roughly 50/50, use extreme value such as 1,000,000 flips.
sns.countplot(x=np.random.binomial(1, 0.5, 1000000));
```

```python
# Outcomes of flipping two coins.
sns.countplot(x=np.random.binomial(2, 0.5, 1000000));
```

```python
# Outcomes of flipping three coins.
sns.countplot(x=np.random.binomial(3, 0.5, 1000000));
```

```python
# Outcomes of flipping fifty coins.
sns.countplot(x=np.random.binomial(50, 0.5, 1000000));
```

### Pauli X gate

In quantum computing, a quantum gate is a quantum circuit that executes a single operation on a quantum register. Quantum gates are the basic building blocks of quantum circuits, in the same way that classical logic gates are the basic building blocks of classical circuits.

One of the most infamous quantum gates is the Pauli X gate, also known as the "bit-flip" gate. This gate flips the state of a qubit from 0 to 1, or vice versa. The Pauli X gate is represented by the matrix:

[0 1]
[1 0]

Another common quantum gate is the Hadamard gate, which puts a qubit into a superposition of the 0 and 1 states. The Hadamard gate is represented by the matrix:

[1  1]
[1 -1]

```python
# Create a circuit.
circuit = qiskit.QuantumCircuit(1, 1)

# Add a gate.
circuit.h(0)

# Measure.
circuit.measure(0, 0)

# Draw the circuit
circuit.draw(output='mpl')
```
